```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Step 1: Create synthetic daily demand data
np.random.seed(42)
days = np.arange(1, 366)  # one year (non-leap)
demand = 50 + 10 * np.sin(2 * np.pi * days / 365) + np.random.normal(0, 2, size=365)

# Step 2: Create a DataFrame
data = pd.DataFrame({'Day': days, 'Demand': demand})

# Add Month column
month_bins = [0, 31, 59, 90, 120, 151, 181, 212, 243, 273, 304, 334, 365]
data['Month'] = pd.cut(data['Day'], bins=month_bins, labels=np.arange(1,13))

# Group by Month and calculate mean demand (fix observed=False warning)
monthly_demand = data.groupby('Month', observed=False)['Demand'].mean().reset_index()

# Step 3: Preprocessing
X = monthly_demand[['Month']].astype(float).values
y = monthly_demand['Demand'].values

scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Step 4: Build Neural Network Model (fix input_shape warning)
model = Sequential([
    Input(shape=(1,)),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# Step 5: Train the model
history = model.fit(X_train, y_train, epochs=300, validation_data=(X_test, y_test), verbose=0)

# Step 6: Evaluate and Predict
y_pred_scaled = model.predict(X_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled)

# Step 7: Plot actual vs predicted demand with month names
plt.figure(figsize=(10,6))
plt.plot(monthly_demand['Month'], monthly_demand['Demand'], 'bo-', label='Actual Monthly Demand')
plt.plot(monthly_demand['Month'], y_pred.flatten(), 'r--', label='Predicted Monthly Demand')

# Define month names
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.xticks(ticks=np.arange(1, 13), labels=month_names)

plt.xlabel('Month')
plt.ylabel('Demand')
plt.title('Monthly Demand Forecasting using Neural Network')
plt.legend()
plt.grid()
plt.show()
```
